```python
from konlpy.tag import Mecab
```

```python
tokenizer = Mecab()
with open("short_clean.txt", "r") as f:
  text = f.read()

tags = tokenizer.pos(text)
tags[:30]
```

```python
pos = 0
for i in range(len(tags)):
  x, y = tags[i]
  if pos + len(x) < len(text) and text[pos+len(x)] == ' ':
    if "EF" in y:
      x = x + "."
      pos -= 1
    x = x + " "
  pos += len(x)
  tags[i] = (x, y)
tags[:30]
```

```python
with_punc = ''.join(x[0] for x in tags)
with_punc
```
